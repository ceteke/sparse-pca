```julia
using Plots
```

```julia
ω = 1.0
τ = 0.5
T = 15
p = 1000
ts = Array{Float64}(0:T*p) / p
Q₀ = 0.2

α₁ = τ*ω*(1+τ/2)
α₂ = τ*(ω-τ/2);
```

```julia
if α₂ == 0
    Q = sqrt.(1 ./ (2*α₁*ts .+ Q₀^(-2)))
else
    Q = sqrt.(α₂ ./ (α₁ .+ (α₂/(Q₀^2) - α₁)*exp.(-2*α₂*ts)))
end;

plot(ts, sqrt.(Q))
```
