```julia
using Random
using Plots
using Statistics
using LinearAlgebra
using Distributions

dtype = Float32;
```

```julia
p = 10000
N = 200
ω = 1
ρ = 0.05

C = randn(dtype, (N, 1))
A = randn(dtype, (N, p))
# as p → ∞ norm(ξ) → √p why?
ξ = rand(Bernoulli(ρ), (1, p)) / sqrt(ρ)

Y = sqrt(ω/p) * C .* ξ + A;
```

```julia
Σ = cov(Y)
plot(diag(Σ))
```

```julia
plot(eigvals(Σ))
```

```julia
ξ * cov(Y) * ξ'
```
